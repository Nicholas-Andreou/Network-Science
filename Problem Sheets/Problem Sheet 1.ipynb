```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.Graph()
e = [(1,2),(1,3),(1,5),(2,3),(2,5),(3,4)]
G.add_edges_from(e)
```

```python
plt.figure()
nx.draw(G, with_labels=True, font_weight='bold')
clust_dic = nx.clustering(G)
avg_clust = list(clust_dic.values())
print(np.mean(avg_clust))
```
