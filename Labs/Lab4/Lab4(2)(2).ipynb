**2)** Consider the configuration model applied to the degree sequence below. Develop a numerical test of our result that the expected number of links between 2 nodes is $k_i k_j/(K-1)$. You do not need to check each distinct pair of nodes, but try a few. 

```python
k = np.array([3, 5, 2, 2, 3, 2, 1, 1, 1]) #Degree sequence
#Add code here
array12 = []
array13 = []
array34 = []
for i in range(1000):
    G = nx.configuration_model(k)
    array12.append(G.number_of_edges(0,1))
    array13.append(G.number_of_edges(0,2))
    array34.append(G.number_of_edges(2,3))
print(f' average number of links between 1-2:{sum(array12)/len(array12)}')
print(f' average number of links between 1-3:{sum(array13)/len(array13)}')
print(f' average number of links between 3-4:{sum(array34)/len(array34)}')
```

```python
def expected_avg(n):
    return (f'expected average of links between {n[0]}-{n[1]}:{(k[n[0]-1]*k[n[1]-1])/(sum(k) - 1)}')

print(expected_avg([1,2]))
print(expected_avg([1,3]))
print(expected_avg([3,4]))
```
