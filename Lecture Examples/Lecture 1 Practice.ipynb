```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge(1,2)
G.edges
```

```python
e = [(1,5),(2,5),(2,3),(3,4),(4,5),(4,6)]
G.add_edges_from(e)
print(G.edges)
print(G.nodes)
```

```python
plt.figure
nx.draw(G, with_labels=True, font_weight='bold')
A = nx.adjacency_matrix(G, nodelist = sorted(G.nodes))
```

```python
A.toarray()
```

```python
print(nx.degree_histogram(G))
Grandom = nx.gnp_random_graph(100,0.05)
nx.draw(Grandom, node_shape='.')
```
